```python
driver = webdriver.Chrome()

driver.get("https://www.casadellibro.com/")

driver.maximize_window()

sleep(2)

#aceptar las cookies
try: 
    driver.find_element("css selector", "#onetrust-accept-btn-handler").click()
except:
    pass

#acceder a libros de ficcion:
driver.find_element("css selector", "#app > div.v-application--wrap > div.when-mobile > div.cabecera.desktop > div.bottom-row > div > div > div:nth-child(2) > a").click()


diccionario_libros = {
    'titulo': [],
    'autor': [],
    'precio': [],
    'editorial': [],
    'idioma': [],
    'num_paginas': []
}

sleep(3)

for child in range(1,6): 
    sleep(3)
#hacer click en producto
    driver.find_element("css selector", f"#buscadorNF > div.col-md-9.col-12 > div > div.grid-view.mt-4 > div:nth-child({child}) > a > div > img").click()

    sleep(2)
    #sacar nombre
    try:
        nombre = driver.find_element("css selector", "#app > div.v-application--wrap > main > div > div > div > div:nth-child(3) > div > div.col-md-5.order-lg-2.col-12.order-1 > div > h1").text
        diccionario_libros['titulo'].append(nombre)
    #print(nombre)
    except:
        desconocido = 'Desconocido'
        diccionario_libros['titulo'].append(desconocido)
#sacar autor
        
    try:
        autor = driver.find_element("css selector", "#app > div.v-application--wrap > main > div > div > div > div:nth-child(3) > div > div.col-md-5.order-lg-2.col-12.order-1 > div > div.text-h5.d-flex.flex-wrap.author.mb-2.justify-center.justify-sm-start > div > a > span:nth-child(1)").text
        #print(autor)
        diccionario_libros['autor'].append(autor)
    except:
        desconocido = 'Desconocido'
        diccionario_libros['autor'].append(desconocido)

    #sacar precio
    try:
        precio = driver.find_element("css selector", "#app > div.v-application--wrap > main > div > div > div > div:nth-child(3) > div > div.border-left.col-md-4.col-12.order-3 > div > div:nth-child(3) > div.col.col-8 > strong").text
        precio = precio.replace(" €", "").replace(",", ".")
        #print(precio)
        diccionario_libros['precio'].append(precio)
    except:
        desconocido = 'Desconocido'
        diccionario_libros['precio'].append(desconocido)

    #sacar editorial
    try:
        editorial = driver.find_element("css selector", "#app > div.v-application--wrap > main > div > div > div > div:nth-child(7) > div > div.caracteristicas.mt-4.pt-4 > div:nth-child(2)").text
        editorial = editorial.replace("Editorial:\n", "")
        #print(editorial)
        diccionario_libros['editorial'].append(editorial)
    except:
        try:
            editorial = driver.find_element("css selector", "#app > div.v-application--wrap > main > div > div > div > div:nth-child(6) > div > div.caracteristicas.mt-4.pt-4 > div:nth-child(2)").text
            editorial = editorial.replace("Editorial:\n", "")
            diccionario_libros['editorial'].append(editorial)
        except:
            desconocido = 'Desconocido'
            diccionario_libros['editorial'].append(desconocido)

    #sacar idioma
    try:
        idioma = driver.find_element("css selector", "#app > div.v-application--wrap > main > div > div > div > div:nth-child(7) > div > div.caracteristicas.mt-4.pt-4 > div:nth-child(3)").text
        idioma = idioma.replace("Idioma:\n", "")
        #print(idioma)
        diccionario_libros['idioma'].append(idioma)
    except:
        try:
            idioma = driver.find_element("css selector", "#app > div.v-application--wrap > main > div > div > div > div:nth-child(6) > div > div.caracteristicas.mt-4.pt-4 > div:nth-child(3)").text
            idioma = idioma.replace("Idioma:\n", "")
            diccionario_libros['idioma'].append(idioma)
        except:
            desconocido = 'Desconocido'
            diccionario_libros['idioma'].append(desconocido)


    #sacar numero paginas
    try:
        paginas = driver.find_element("css selector", "#app > div.v-application--wrap > main > div > div > div > div:nth-child(7) > div > div.caracteristicas.mt-4.pt-4 > div:nth-child(1)").text
        paginas = paginas.replace("Nº de páginas:\n", "")
        #print(paginas)
        diccionario_libros['num_paginas'].append(paginas)
    except: 
        try:
            paginas = driver.find_element("css selector", "#app > div.v-application--wrap > main > div > div > div > div:nth-child(6) > div > div.caracteristicas.mt-4.pt-4 > div:nth-child(1)").text
            paginas = paginas.replace("Nº de páginas:\n", "")
            diccionario_libros['num_paginas'].append(paginas)
        except:
            desconocido = 'Desconocido'
            diccionario_libros['num_paginas'].append(desconocido)


    driver.back()

print(diccionario_libros)
```

```python
lista_libros = list(zip(diccionario_libros['titulo'], diccionario_libros['autor'], diccionario_libros['precio'], diccionario_libros['editorial'], diccionario_libros['idioma'], diccionario_libros['num_paginas']))
```

```python
df_libros_ficcion = pd.DataFrame(lista_libros, columns=['Titulo', 'Autor', 'Precio', 'Editorial', 'Idioma', 'N.Páginas'])
```
